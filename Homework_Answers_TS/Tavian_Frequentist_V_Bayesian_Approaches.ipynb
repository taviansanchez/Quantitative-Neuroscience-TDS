```python
# Exercise 1: Demonstrating statistical significance of a positive test

# Parameters
false_positive_rate = 0.05  # probability a healthy person tests positive

# Under the null hypothesis: nobody has HIV
# So any positive result is a false positive
p_positive_by_chance = false_positive_rate

print(f"Probability of a positive test by chance: {p_positive_by_chance:.2f}")

# Check if it's statistically significant at p < 0.05
if p_positive_by_chance < 0.05:
    print("The positive test IS statistically significant (p < 0.05).")
else:
    print("The positive test is NOT statistically significant (p >= 0.05).")
```

```python
# Parameters
false_positive_rate = 0.05  # probability a healthy person tests positive
false_negative_rate = 0.0   # probability an infected person tests negative
infection_rates = [i/10 for i in range(11)]  # priors from 0 to 1 in steps of 0.1

# Calculate posterior probability for each prior
posterior_probs = []
for prior in infection_rates:
    P_positive = (1 - false_negative_rate) * prior + false_positive_rate * (1 - prior)
    posterior = ((1 - false_negative_rate) * prior) / P_positive
    posterior_probs.append(posterior)

# Print results
for prior, posterior in zip(infection_rates, posterior_probs):
    print(f"Prior infection rate: {prior:.1f}, Probability infected given positive test: {posterior:.2f}")
```

```python
import matplotlib.pyplot as plt

# Parameters
P_data_given_hypothesis = 1.0       # analogous to 1 - false negative rate
P_data_given_not_hypothesis = 0.05  # analogous to false positive rate
priors = [i/10 for i in range(11)]  # prior probability of hypothesis from 0 to 1

# Calculate posterior probabilities
posterior_probs = []
for prior in priors:
    P_data = (P_data_given_hypothesis * prior +
              P_data_given_not_hypothesis * (1 - prior))
    posterior = (P_data_given_hypothesis * prior) / P_data
    posterior_probs.append(posterior)

# Plot
plt.figure(figsize=(8,5))
plt.plot(priors, posterior_probs, marker='o', color='blue', linewidth=2)
plt.title("Posterior Probability vs Prior Probability")
plt.xlabel("Prior Probability of Hypothesis")
plt.ylabel("Posterior Probability Given Data")
plt.grid(True)
plt.show()
```
