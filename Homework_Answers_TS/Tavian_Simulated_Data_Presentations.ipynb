```python
from statsmodels.stats.power import TTestIndPower
import numpy as np

def noninferiority_power_analysis(
    sd_iso,
    sd_cd8,
    delta,
    alpha=0.05,
    power=0.80,
    n_per_group_observed=None
):
    """
    Non-inferiority power analysis for a two-group continuous outcome.

    Parameters
    ----------
    sd_iso : float
        Standard deviation in the iso group.
    sd_cd8 : float
        Standard deviation in the α-CD8 group.
    delta : float
        Non-inferiority margin (absolute units of the metric, e.g. 10% of iso mean).
        This is the largest loss you consider biologically acceptable.
    alpha : float
        One-sided significance level (typically 0.05 or 0.025).
    power : float
        Desired power (usually 0.8).
    n_per_group_observed : int, optional
        If provided, compute achieved power at this sample size.
    """
    analysis = TTestIndPower()

    # 1) pooled SD
    sigma_pooled = np.sqrt((sd_iso**2 + sd_cd8**2) / 2)

    # 2) standardized effect size at the NI margin
    effect_size = delta / sigma_pooled

    print(f"Pooled SD: {sigma_pooled:.3f}")
    print(f"Effect size at NI margin (d = Δ/σ): {effect_size:.3f}")

    # 3) Required N per group for the chosen power
    n_required = analysis.solve_power(
        effect_size=effect_size,
        alpha=alpha,
        power=power,
        alternative='larger'  # one-sided: H1 is 'greater than' the NI boundary
    )
    print(f"Required N per group for {power*100:.0f}% power: {n_required:.1f}")

    # 4) Optional: achieved power at your actual N
    if n_per_group_observed is not None:
        achieved_power = analysis.solve_power(
            effect_size=effect_size,
            alpha=alpha,
            nobs1=n_per_group_observed,
            alternative='larger'
        )
        print(f"Achieved power at N = {n_per_group_observed} per group: {achieved_power:.3f}")

    return n_required
```

```python
# If needed:
# !pip install statsmodels

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.power import TTestIndPower

# ---------- 1. Simulate branches data ----------

def simulate_branches(n_per_group=100,
                      group_labels=('iso', 'α-CD8'),
                      mean_values=(57, 60),
                      std_values=(12.6, 13.9),
                      seed=42):
    """
    Simulate 'number of branches' data for iso vs α-CD8 groups.
    Returns a DataFrame with columns: Group, Value.
    """
    if seed is not None:
        np.random.seed(seed)

    data = []
    for group, mean, std in zip(group_labels, mean_values, std_values):
        y = np.random.normal(mean, std, n_per_group)
        data.extend([(group, val) for val in y])

    df = pd.DataFrame(data, columns=['Group', 'Value'])
    return df


# ---------- 2. Non-inferiority power analysis + pretty outputs ----------

def noninferiority_power_report_for_branches(
    n_per_group_microglia=100,
    mean_values=(57, 60),
    std_values=(12.6, 13.9),
    delta_percent=0.20,   # 20% NI margin
    alpha=0.025,          # one-sided alpha
    power_target=0.80,
    seed=42,
    make_power_curve=True
):
    """
    Non-inferiority power analysis for 'number of branches' with:
      - Δ = delta_percent * iso mean
      - one-sided alpha = alpha
      - target power = power_target

    Produces:
      - a nicely formatted pandas table of results
      - an optional power curve plot (N mice per group vs. power)
    """

    # 1) Simulate data (to match your plotting distribution)
    df = simulate_branches(
        n_per_group=n_per_group_microglia,
        mean_values=mean_values,
        std_values=std_values,
        seed=seed
    )

    iso_vals = df[df['Group'] == 'iso']['Value']
    cd8_vals = df[df['Group'] == 'α-CD8']['Value']

    mean_iso = iso_vals.mean()
    mean_cd8 = cd8_vals.mean()
    sd_iso = iso_vals.std(ddof=1)
    sd_cd8 = cd8_vals.std(ddof=1)

    # 2) Non-inferiority margin: 20% of iso mean
    delta = delta_percent * mean_iso

    # 3) Pooled SD and effect size
    sigma_pooled = np.sqrt((sd_iso**2 + sd_cd8**2) / 2)
    effect_size = delta / sigma_pooled   # d = Δ / σ_pooled

    # 4) Power analysis (one-sided NI)
    analysis = TTestIndPower()

    # Required N (independent units, i.e. mice) per group
    n_required_mice = analysis.solve_power(
        effect_size=effect_size,
        alpha=alpha,
        power=power_target,
        alternative='larger'
    )

    # Achieved power at your simulated microglia N (just for display)
    achieved_power_microglia = analysis.solve_power(
        effect_size=effect_size,
        alpha=alpha,
        nobs1=n_per_group_microglia,
        alternative='larger'
    )

    # ---------- 5. Build a presentation-ready summary table ----------

    summary = pd.DataFrame({
        "Metric": ["Number of branches"],
        "iso mean ± SD": [f"{mean_iso:.2f} ± {sd_iso:.2f}"],
        "α-CD8 mean ± SD": [f"{mean_cd8:.2f} ± {sd_cd8:.2f}"],
        "NI margin Δ (20% of iso mean)": [f"{delta:.2f} branches"],
        "Effect size d (Δ / σ_pooled)": [f"{effect_size:.2f}"],
        "Target power": [f"{power_target*100:.0f}%"],
        "α (one-sided)": [f"{alpha:.3f}"],
        "Required N / group (mice)": [f"{n_required_mice:.1f}"],
        f"Power at N={n_per_group_microglia} (microglia/group)": [f"{achieved_power_microglia:.3f}"]
    })

    print("\nNon-inferiority power analysis: Number of Branches\n")
    display(summary)

    print("\nInterpretation:")
    print(f"- Δ is set to 20% of the iso mean, i.e. a {delta:.2f}-branch loss.")
    print("- N in the table refers to independent experimental units (mice),")
    print("  not individual microglia (which are nested within mice).")

    # ---------- 6. Optional: Power curve figure (N mice/group vs power) ----------

    if make_power_curve:
        n_range = np.arange(5, 51)  # 5 to 50 mice per group
        powers = [
            analysis.solve_power(
                effect_size=effect_size,
                alpha=alpha,
                nobs1=n,
                alternative='larger'
            )
            for n in n_range
        ]

        plt.figure(figsize=(6, 4))
        plt.plot(n_range, powers)
        plt.axhline(power_target, linestyle='--')
        plt.xlabel("N mice per group")
        plt.ylabel("Power")
        plt.title("Non-inferiority power curve\n(Number of branches, Δ = 20% of iso mean)")
        plt.tight_layout()
        plt.show()


# ---- Run it for your branches metric with Δ = 20% and α = 0.025 ----
noninferiority_power_report_for_branches()
```

\]

### **Required sample size per group (mice)**  
(approximate analytic form solved numerically by `statsmodels`)
\[
n \approx 
\frac{2\,\left(z_{1-\alpha} + z_{1-\beta}\right)^2}{d^2}
\]

---

**Where:**  
- \(\alpha = 0.025\) (one-sided)  
- \(1-\beta = 0.80\) (power)  
- \(d = \Delta / \sigma_{\text{pooled}}\)

```python
import numpy as np
import pandas as pd
from scipy import stats

def get_mouse_means(df, group_col='Group', mouse_col='MouseID', value_col='CellValue'):
    """
    Collapse per-microglia data to per-mouse means.
    Returns iso_mouse, cd8_mouse (1D arrays of mouse-level means).
    """
    mouse_means = (
        df.groupby([group_col, mouse_col])[value_col]
        .mean()
        .reset_index()
    )
    
    iso = mouse_means[mouse_means[group_col] == 'iso'][value_col].values
    cd8 = mouse_means[mouse_means[group_col] == 'α-CD8'][value_col].values
    
    return iso, cd8, mouse_means

def noninferiority_CI_and_welch(
    df_branches,
    delta_percent=0.20,   # 20% NI margin (change as needed)
    alpha=0.025           # one-sided alpha for NI
):
    """
    Non-inferiority analysis for 'number of branches' at the mouse level:
      - CI-based NI check
      - one-sided Welch NI t-test
    
    Assumes:
      * lower values = worse (inferior morphology)
      * Δ is the maximum tolerated LOSS (iso - αCD8) in the α-CD8 group
    """
    iso_mouse, cd8_mouse, mouse_means = get_mouse_means(df_branches)
    
    n_iso  = len(iso_mouse)
    n_cd8  = len(cd8_mouse)
    
    mean_iso = iso_mouse.mean()
    mean_cd8 = cd8_mouse.mean()
    
    # Difference defined as α-CD8 minus iso
    diff = mean_cd8 - mean_iso
    
    sd_iso = iso_mouse.std(ddof=1)
    sd_cd8 = cd8_mouse.std(ddof=1)
    
    # Non-inferiority margin: Δ = (delta_percent * iso mean)
    Delta = delta_percent * mean_iso   # positive number (e.g. 20% of iso mean)
    
    # Welch SE for difference (cd8 - iso)
    se_diff = np.sqrt(sd_cd8**2 / n_cd8 + sd_iso**2 / n_iso)
    
    # Welch degrees of freedom
    df_num = (sd_cd8**2 / n_cd8 + sd_iso**2 / n_iso)**2
    df_den = (sd_cd8**2 / n_cd8)**2 / (n_cd8 - 1) + (sd_iso**2 / n_iso)**2 / (n_iso - 1)
    df_welch = df_num / df_den
    
    # ---------- 1) CI-BASED NON-INFERIORITY ----------
    # We want CI for (mean_cd8 - mean_iso).
    # NI if the lower bound > -Delta  (i.e. we can rule out losses >= Delta)
    
    # Use a (1 - 2*alpha)*100% two-sided CI (e.g. alpha=0.025 -> 95% CI)
    ci_level = 1 - 2*alpha
    t_crit = stats.t.ppf(1 - (1 - ci_level)/2, df_welch)
    
    ci_lower = diff - t_crit * se_diff
    ci_upper = diff + t_crit * se_diff
    
    ni_by_ci = ci_lower > -Delta
    
    # ---------- 2) WELCH ONE-SIDED NI t-TEST ----------
    # H0: μ_cd8 - μ_iso <= -Delta  vs  H1: μ_cd8 - μ_iso > -Delta
    # Shift the difference by +Delta and test > 0
    shifted_diff = diff + Delta
    t_stat = shifted_diff / se_diff
    p_one_sided = 1 - stats.t.cdf(t_stat, df_welch)
    
    ni_by_ttest = p_one_sided < alpha
    
    # ---------- PRETTY READOUT ----------
    print("Non-inferiority analysis: Number of branches (mouse level)")
    print("-----------------------------------------------------------")
    print(f"N mice (iso)   = {n_iso}")
    print(f"N mice (α-CD8) = {n_cd8}")
    print()
    print(f"iso mean ± SD      = {mean_iso:.2f} ± {sd_iso:.2f}")
    print(f"α-CD8 mean ± SD    = {mean_cd8:.2f} ± {sd_cd8:.2f}")
    print(f"Difference (α-CD8 − iso) = {diff:.2f} branches")
    print()
    print(f"Non-inferiority margin Δ ({delta_percent*100:.0f}% of iso mean):")
    print(f"  Δ = {Delta:.2f} branches (maximum tolerated loss)")
    print()
    print(f"{ci_level*100:.1f}% CI for (α-CD8 − iso):")
    print(f"  [{ci_lower:.2f}, {ci_upper:.2f}] branches")
    print(f"CI-based NI conclusion (lower bound > −Δ):  {ni_by_ci}")
    print()
    print("Welch one-sided NI t-test:")
    print("  H0: μ_cd8 − μ_iso ≤ −Δ")
    print("  H1: μ_cd8 − μ_iso > −Δ")
    print(f"  t = {t_stat:.3f},  df ≈ {df_welch:.1f},  p (one-sided) = {p_one_sided:.4f}")
    print(f"T-test NI conclusion (p < {alpha}):          {ni_by_ttest}")
    print("-----------------------------------------------------------")
    print("Note: microglia are nested within mice; this analysis uses")
    print("      one value per mouse (mean branches per mouse).")

    return {
        "mean_iso": mean_iso,
        "mean_cd8": mean_cd8,
        "diff": diff,
        "Delta": Delta,
        "ci_lower": ci_lower,
        "ci_upper": ci_upper,
        "ni_by_ci": ni_by_ci,
        "t_stat": t_stat,
        "df_welch": df_welch,
        "p_one_sided": p_one_sided,
        "ni_by_ttest": ni_by_ttest
    }
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# ============================================
# 1. GENERIC SIMULATOR: microglia within mice
# ============================================

def simulate_metric_data(
    n_mice_per_group=22,
    cells_per_mouse=8,
    group_labels=('iso', 'α-CD8'),
    mean_values=(0.0, 0.0),
    std_values=(1.0, 1.0),
    value_col='Value',
    seed=42,
):
    """
    Simulate a metric for microglia nested within mice in two groups.
    Returns a DataFrame with columns: Group, MouseID, <value_col>.
    """
    if seed is not None:
        np.random.seed(seed)

    records = []
    for group, mean, std in zip(group_labels, mean_values, std_values):
        for m in range(n_mice_per_group):
            mouse_id = f"{group}_mouse{m+1}"
            y = np.random.normal(mean, std, cells_per_mouse)
            for val in y:
                records.append((group, mouse_id, val))

    df = pd.DataFrame(records, columns=['Group', 'MouseID', value_col])
    return df


# ============================================
# 2. GENERIC PLOTTER: grey cells, black mice
# ============================================

def plot_metric_data(
    df,
    value_col='Value',
    ylabel='Metric',
    title='Simulated Data',
    ylim=None,
):
    """
    Plot all microglia (grey), per-mouse means (black dots),
    and group mean bars (black).
    """
    plt.figure(figsize=(4, 5))

    # a) All microglia as grey dots
    sns.stripplot(
        data=df,
        x='Group',
        y=value_col,
        color='0.7',     # grey
        size=4,
        jitter=0.25,
        alpha=0.7
    )

    # b) Per-mouse means as larger black dots
    mouse_means = (
        df.groupby(['Group', 'MouseID'])[value_col]
        .mean()
        .reset_index()
    )

    sns.stripplot(
        data=mouse_means,
        x='Group',
        y=value_col,
        dodge=False,
        color='black',
        size=7,
        jitter=False
    )

    # c) Group mean bars (based on mouse means)
    group_means = mouse_means.groupby('Group')[value_col].mean()
    for i, group in enumerate(group_means.index):
        gm = group_means.loc[group]
        plt.plot([i - 0.25, i + 0.25], [gm, gm],
                 color='black', lw=2.5, solid_capstyle='round')

    plt.ylabel(ylabel)
    plt.xlabel('')
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    sns.despine()
    plt.tight_layout()
    plt.show()

    return mouse_means  # useful for checking mouse-level data


# ============================================
# 3. NON-INFERIORITY (CI + one-sided Welch)
# ============================================

def get_mouse_means(df, group_col='Group', mouse_col='MouseID', value_col='Value'):
    mouse_means = (
        df.groupby([group_col, mouse_col])[value_col]
        .mean()
        .reset_index()
    )
    iso = mouse_means[mouse_means[group_col] == 'iso'][value_col].values
    cd8 = mouse_means[mouse_means[group_col] == 'α-CD8'][value_col].values
    return iso, cd8, mouse_means


def noninferiority_CI_and_welch(
    df,
    value_col='Value',
    metric_name='Metric',
    delta_percent=0.20,   # 20% NI margin
    alpha=0.025           # one-sided NI alpha
):
    """
    Computes:
      (1) CI-based NI decision
      (2) one-sided NI Welch t-test
    """
    iso, cd8, mouse_means = get_mouse_means(df, value_col=value_col)

    n_iso = len(iso)
    n_cd8 = len(cd8)

    mean_iso = iso.mean()
    mean_cd8 = cd8.mean()
    diff = mean_cd8 - mean_iso   # α-CD8 − iso

    sd_iso = iso.std(ddof=1)
    sd_cd8 = cd8.std(ddof=1)

    # NI margin: allowed loss relative to iso
    Delta = delta_percent * mean_iso

    # SE and df for Welch
    se_diff = np.sqrt(sd_cd8**2 / n_cd8 + sd_iso**2 / n_iso)
    df_num = (sd_cd8**2 / n_cd8 + sd_iso**2 / n_iso)**2
    df_den = ((sd_cd8**2 / n_cd8)**2) / (n_cd8 - 1) + ((sd_iso**2 / n_iso)**2) / (n_iso - 1)
    df_welch = df_num / df_den

    # 1) CI-based NI
    ci_level = 1 - 2*alpha          # e.g. 95% for alpha=0.025
    t_crit = stats.t.ppf(1 - (1 - ci_level)/2, df_welch)
    ci_lower = diff - t_crit * se_diff
    ci_upper = diff + t_crit * se_diff
    ni_by_ci = ci_lower > -Delta

    # 2) One-sided NI Welch
    shifted_diff = diff + Delta
    t_stat = shifted_diff / se_diff
    p_one_sided = 1 - stats.t.cdf(t_stat, df_welch)
    ni_by_ttest = p_one_sided < alpha

    # Pretty print
    print(f"\nNON-INFERIORITY ANALYSIS: {metric_name}")
    print("==============================================")
    print(f"N iso   = {n_iso}")
    print(f"N α-CD8 = {n_cd8}")
    print()
    print(f"iso mean ± SD      = {mean_iso:.2f} ± {sd_iso:.2f}")
    print(f"α-CD8 mean ± SD    = {mean_cd8:.2f} ± {sd_cd8:.2f}")
    print(f"Difference (α-CD8 − iso) = {diff:.2f}")
    print()
    print(f"NI margin Δ = {Delta:.2f} ({delta_percent*100:.0f}% of iso mean)")
    print(f"{ci_level*100:.1f}% CI = [{ci_lower:.2f}, {ci_upper:.2f}]")
    print(f"NI by CI (CI_lower > -Δ): {ni_by_ci}")
    print()
    print("One-sided NI Welch test:")
    print(f"t = {t_stat:.3f}, df = {df_welch:.1f}, p = {p_one_sided:.5f}")
    print(f"NI by t-test (p < {alpha}): {ni_by_ttest}")
    print("==============================================\n")

    return {
        "mean_iso": mean_iso,
        "mean_cd8": mean_cd8,
        "diff": diff,
        "Delta": Delta,
        "ci_lower": ci_lower,
        "ci_upper": ci_upper,
        "ni_by_ci": ni_by_ci,
        "t_stat": t_stat,
        "df_welch": df_welch,
        "p_one_sided": p_one_sided,
        "ni_by_ttest": ni_by_ttest,
        "mouse_means": mouse_means
    }


# ============================================
# 4. RUN FOR ALL METRICS
#    (tweak means/SDs/ylims as desired)
# ============================================

# 1) Iba1+ soma size (µm^2)
df_soma = simulate_metric_data(
    mean_values=(34.6, 37.1),
    std_values=(8.2, 7.6),
    value_col='SomaSize'
)
plot_metric_data(
    df_soma,
    value_col='SomaSize',
    ylabel='Cell Body Size (µm²)',
    title='Simulated Iba1⁺ Cell Body Size',
    ylim=(0, 70)
)
res_soma = noninferiority_CI_and_welch(
    df_soma,
    value_col='SomaSize',
    metric_name='Iba1⁺ soma size'
)

# 2) Number of branches
df_branches = simulate_metric_data(
    mean_values=(57, 60),
    std_values=(12.6, 13.9),
    value_col='Branches'
)
plot_metric_data(
    df_branches,
    value_col='Branches',
    ylabel='Number of Branches',
    title='Simulated Branching',
    ylim=(0, 115)
)
res_branches = noninferiority_CI_and_welch(
    df_branches,
    value_col='Branches',
    metric_name='Number of branches'
)

# 3) Average branch length (µm)
df_branch_len = simulate_metric_data(
    mean_values=(2.74, 2.67),
    std_values=(0.6, 0.9),
    value_col='BranchLen'
)
plot_metric_data(
    df_branch_len,
    value_col='BranchLen',
    ylabel='Average Branch Length (µm)',
    title='Simulated Average Branch Length',
    ylim=(0, 5)
)
res_branch_len = noninferiority_CI_and_welch(
    df_branch_len,
    value_col='BranchLen',
    metric_name='Average branch length'
)

# 4) Number of junctions
df_junctions = simulate_metric_data(
    mean_values=(28, 35),
    std_values=(8.0, 8.5),
    value_col='Junctions'
)
plot_metric_data(
    df_junctions,
    value_col='Junctions',
    ylabel='Number of Junctions',
    title='Simulated Junctional Data',
    ylim=(0, 60)
)
res_junctions = noninferiority_CI_and_welch(
    df_junctions,
    value_col='Junctions',
    metric_name='Number of junctions'
)

# 5) Longest − shortest branch length (µm)
df_longshort = simulate_metric_data(
    mean_values=(40, 38),
    std_values=(7.6, 7.9),
    value_col='LongShort'
)
plot_metric_data(
    df_longshort,
    value_col='LongShort',
    ylabel='Longest − Shortest Branch (µm)',
    title='Simulated Length Difference',
    ylim=(0, 80)
)
res_longshort = noninferiority_CI_and_welch(
    df_longshort,
    value_col='LongShort',
    metric_name='Longest − shortest branch length'
)

# 6) Number of endpoints (choose plausible values; adjust as needed)
df_endpoints = simulate_metric_data(
    mean_values=(65, 70),
    std_values=(15, 16),
    value_col='Endpoints'
)
plot_metric_data(
    df_endpoints,
    value_col='Endpoints',
    ylabel='Number of Endpoints',
    title='Simulated Endpoint Data',
    ylim=(0, 130)
)
res_endpoints = noninferiority_CI_and_welch(
    df_endpoints,
    value_col='Endpoints',
    metric_name='Number of endpoints'
)
```

```python
def plot_NI_forest_single(res, metric_name='Metric',
                          x_label='α-CD8 − iso',
                          ax=None):
    """
    Forest-style plot for a single metric:
      - point = mean difference
      - horizontal line = CI
      - vertical dashed line at -Δ (NI boundary)
      - vertical solid line at 0 (no difference)
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=(3, 4))

    diff = res['diff']
    ci_lower = res['ci_lower']
    ci_upper = res['ci_upper']
    Delta = res['Delta']

    # CI as horizontal line at y=0
    ax.hlines(y=0, xmin=ci_lower, xmax=ci_upper, lw=2, color='black')
    # Point estimate
    ax.plot(diff, 0, 'o', color='black', ms=6)

    # NI boundary at -Δ
    ax.axvline(-Delta, linestyle='--', color='red', lw=1.5, label='NI boundary (−Δ)')
    # Zero line
    ax.axvline(0, linestyle='-', color='gray', lw=1, alpha=0.7)

    ax.set_xlabel(x_label)
    ax.set_yticks([])
    ax.set_title(f'NI result: {metric_name}', fontsize=10)

    # Nice x-limits
    span = max(abs(ci_lower), abs(ci_upper), abs(Delta)) * 1.2
    ax.set_xlim(-span, span)

    # Label NI boundary
    ax.text(-Delta, 0.1, '−Δ', color='red', ha='center', va='bottom', fontsize=9)

    sns.despine(left=True, ax=ax)
    return ax
```

```python
def scatter_nested_on_ax(ax, df, value_col, ylabel, title, ylim=None):
    """
    Panel A: grey microglia, black mouse means, black group mean bar.
    """
    # all microglia
    sns.stripplot(
        data=df,
        x='Group',
        y=value_col,
        color='0.7',
        size=3,
        jitter=0.25,
        alpha=0.7,
        ax=ax
    )

    # per-mouse means
    mouse_means = (
        df.groupby(['Group', 'MouseID'])[value_col]
        .mean()
        .reset_index()
    )
    sns.stripplot(
        data=mouse_means,
        x='Group',
        y=value_col,
        color='black',
        size=6,
        jitter=False,
        ax=ax
    )

    # group mean bars
    group_means = mouse_means.groupby('Group')[value_col].mean()
    for i, group in enumerate(group_means.index):
        gm = group_means.loc[group]
        ax.plot([i - 0.25, i + 0.25], [gm, gm],
                color='black', lw=2.5, solid_capstyle='round')

    ax.set_ylabel(ylabel)
    ax.set_xlabel('')
    ax.set_title(title, fontsize=10)
    if ylim is not None:
        ax.set_ylim(*ylim)
    sns.despine(ax=ax)
    return mouse_means

def make_AB_figure(df, value_col, ylabel, title, metric_name,
                   ylim=None, delta_percent=0.20, alpha=0.025):
    """
    Makes a two-panel figure:
      A: scatter plot (microglia + mouse means)
      B: NI forest plot (difference ± CI vs NI boundary)
    """
    # run NI (quietly)
    res = noninferiority_CI_and_welch(
        df,
        value_col=value_col,
        metric_name=metric_name,
        delta_percent=delta_percent,
        alpha=alpha,
        verbose=False
    )

    fig, (axA, axB) = plt.subplots(
        1, 2,
        figsize=(7, 4),
        gridspec_kw={'width_ratios': [2.2, 1]}
    )

    scatter_nested_on_ax(axA, df, value_col, ylabel, title, ylim)
    plot_NI_forest_single(res, metric_name=metric_name, ax=axB)

    fig.suptitle(metric_name, fontsize=12, y=1.02)
    plt.tight_layout()
    return fig, res
```
