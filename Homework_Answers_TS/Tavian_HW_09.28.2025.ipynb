Exercises: #1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = {
    "Age": [3,4,5,6,7,8,9,11,12,14,15,16,17],
    "Wing_Length": [1.4,1.5,2.2,2.4,3.1,3.2,3.2,3.9,4.1,4.7,4.5,5.2,5.0]
}
df = pd.DataFrame(data)

# Scatter plot
sns.scatterplot(x="Age", y="Wing_Length", data=df, s=80)
plt.title("Age vs Wing Length")
plt.xlabel("Age")
plt.ylabel("Wing Length")
plt.show()
```

Exercises: #2

```python
import statsmodels.api as sm
import numpy as np

# Define variables
X = sm.add_constant(df["Age"])  # adds intercept
y = df["Wing_Length"]

# Fit regression
model = sm.OLS(y, X).fit()

# Predicted values
df["Predicted"] = model.predict(X)

# Plot scatter + regression line
plt.scatter(df["Age"], df["Wing_Length"], label="Data", s=80)
plt.plot(df["Age"], df["Predicted"], color="red", label="Regression Line")
plt.xlabel("Age")
plt.ylabel("Wing Length")
plt.title("Regression Line: Age vs Wing Length")
plt.legend()
plt.show()
```

Exercises: #3

```python
from scipy.stats import linregress

# Run linear regression with SciPy
slope, intercept, r_value, p_value, std_err = linregress(df["Age"], df["Wing_Length"])

print(f"Slope = {slope:.3f}")
print(f"Intercept = {intercept:.3f}")
print(f"p-value for slope = {p_value:.5f}")
print(f"R-squared = {r_value**2:.3f}")
```

A different way to do Exercise 3 from ChatGPT

```python
# Perform a t-test on the slope (Age coefficient)
t_test = model.t_test("Age = 0")
print(t_test)
```

Exercises: #4

```python
# Get 95% confidence intervals
conf_int = model.conf_int(alpha=0.05)
print("95% Confidence Intervals:\n", conf_int)

# Extract slope CI
slope_ci = conf_int.loc["Age"]

# Plot slope with CI
sns.regplot(x="Age", y="Wing_Length", data=df, ci=95, line_kws={"color":"red"})
plt.title("Regression Line with 95% Confidence Interval")
plt.show()
```

Exercises: #5

```python
r_squared = model.rsquared
print("R-squared:", r_squared)
```

Exercises: #6

```python
from scipy import stats

r, p = stats.pearsonr(df["Age"], df["Wing_Length"])
print(f"Pearson's r = {r:.3f}, p-value = {p:.3f}")
```

Exercises: #7

```python
# Add noise to Wing Length
np.random.seed(42)
df["Wing_Length_Noisy"] = df["Wing_Length"] + np.random.normal(0, 0.5, len(df))

# Refit regression with noisy data
X_noisy = sm.add_constant(df["Age"])
y_noisy = df["Wing_Length_Noisy"]
model_noisy = sm.OLS(y_noisy, X_noisy).fit()

# Plot original vs noisy regression
plt.scatter(df["Age"], df["Wing_Length_Noisy"], label="Noisy Data", s=80, alpha=0.6)
plt.plot(df["Age"], model_noisy.predict(X_noisy), color="red", label="Noisy Regression Line")
plt.xlabel("Age")
plt.ylabel("Wing Length (Noisy)")
plt.title("Regression with Added Noise")
plt.legend()
plt.show()

print(model_noisy.summary())
```
